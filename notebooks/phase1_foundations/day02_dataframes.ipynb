```python
data = [(1, "Alice", 29), (2, "Ben", 34), (3, "Cara", 41)]
columns = ["id", "name", "age"]
df = spark.createDataFrame(data, columns)
df.display()
```

```python
#Rendered table view (Databricks-specific)
df.display()
```

```python
#Plain text table (works in any Spark environment)
df.show()
```
