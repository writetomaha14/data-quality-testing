```python
data = [(1, "Alice", "IT", 29, 55000),
        (2, "Ben", "HR", 34, 48000),
        (3, "Cara", "IT", 41, 62000),
        (4, "Dev", "Finance", 26, 45000),
        (5, "Ella", None, 31, 51000)]
columns = ["id", "name", "department", "age", "salary"]
df = spark.createDataFrame(data, columns)
df.display()
```

```python
#Selecting one or more column
df.select("name").display()
df.select("name", "department", "salary").display()
```

```python
#Selecting all the columns
df.select("*").display()
```

```python
#Selecting the columns with expressions
from pyspark.sql.functions import col
df.select(col("name"), col("salary")).display()
```

```python
# Filtering or Choosing Rows
#filter()and where() do exactly the same thing in Pyspark - they're aliases. use whichever reads natural. where() feel familiar coming from SQL

df.filter(df.department=="IT").display()
df.where(df.department=="IT").display()
```

```python
#Comparison Operators
df.filter(df.department=="HR").display()
df.filter(df.department!="HR").display()
df.filter(df.salary >"50000").display()
df.filter(df.salary <"50000").display()
df.filter(df.salary >="50000").display()
df.filter(df.salary <="50000").display()
```

```python
#Combining Conditions AND /OR
#And
df.filter((df.department=="IT") & (df.salary=="55000")).display()

#OR
df.filter((df.department=="IT") | (df.salary>="55000")).display()
```

```python
# The isin() method (Psparks 'IN')
#Equivalent to SQL's WHERE department ('IN' ('IT', 'HR'))
df.filter(df.department.isin(['IT', 'HR','Finance'])).display()

#The opposite of isin() is the isnot() method
#Equivalent to SQL's)
df.filter(~df.department.isin(['IT', 'HR','Finance'])).display()
```

```python
#The between() method
#Equivalent to SQL's Between -inclusive on both ends
df.filter(df.age.between(25,35)).display()

#The opposite of between() is the notbetween() method
#Equivalent to SQL's NOT BETWEEN -exclusive on both ends
df.filter(~df.age.between(25,35)).display()
```

```python
# The Like Method
#Equivalent to SQL's LIKE
df.filter(df.name.like("A%")).display()

```

```python
# The Not Like Method
#Equivalent to SQL's NOT LIKE
df.filter(~df.name.like("A%")).display()
```

```python
# NULL - a special cCase (important, same trap as SQL)
# Just like SQL, PySpark's nulls are not zero or empty string, and == does not work for finding them.

#Wrong way: df.filter(df.department == None) — this does not work.

# Find the row where department is missing
df.filter(df.department.isNull()).display()
```

```python
# Find the row where department is not missing
df.filter(df.department.isNotNull()).display()
```

```python
# Putting It All Together — A Realistic Example
# Table: orders (order_id, customer_id, order_date, status, amount, region)

orders_data = [
    (1001, 501, "2026-07-10", "completed", 1500.00, "North"),
    (1002, 502, "2026-07-11", "failed", 2200.00, "South"),
    (1003, 503, "2026-07-12", "completed", 850.00, "East"),
    (1004, None, "2026-07-12", "failed", 1800.00, "West"),
    (1005, 504, "2026-07-13", "pending", 950.00, "North"),
    (1006, 505, "2026-07-13", "failed", 3200.00, "South"),
    (1007, 506, "2026-07-14", "completed", 600.00, "East"),
    (1008, 507, "2026-07-15", "failed", 1250.00, "North"),
    (1009, 508, "2026-07-15", "completed", 4100.00, "West"),
    (1010, None, "2026-07-16", "pending", 750.00, "South")
]

orders_columns = ["order_id", "customer_id", "order_date", "status", "amount", "region"]
orders = spark.createDataFrame(orders_data, orders_columns)

print("All orders:")
orders.display()
```

```python
# Filter for failed orders over $1000 with valid customer_id
from pyspark.sql.functions import col

result = orders.select("order_id", "amount", "region") \
    .filter((col("status") == "failed") &
            (col("amount") > 1000) &
            (col("customer_id").isNotNull()))

print("Failed orders > $1000 with customer_id:")
result.display()
```
